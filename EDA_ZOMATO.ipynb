```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('zomato.csv', encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
# missing values
df.isnull().sum()
```

```python
# what is inputation ?--process of treating missing values
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val,labels=country_names,autopct='%1.2f%%')
```

```python
## pie chart of top 5 companies that uses zomato 
plt.pie(country_val[:5],labels=country_names[:5],autopct='%1.2f%%')
```

```python
plt.pie(country_val[:10],labels=country_names[:10],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
sns.countplot(x="Rating color",data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'black'])
```
